```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
cv2.__version__
```

```python
Hsmt_tyrod = cv2.imread('C:/Users/sheim/Documents/REU-DataScienceProgram/REU-Python/image_data_analisys/thyr.jpg')

thyroid2pic = cv2.imread('C:/Users/sheim/Documents/REU-DataScienceProgram/REU-Python/image_data_analisys/thyroid2pic.jpg')



print('Showing size Hsmt_tyrod', Hsmt_tyrod.shape)
print('Showing size Hsmt_tyrod', thyroid2pic.shape)
#verify the dimention of the matrix

Hsmt_tyrod_reshaped = cv2.resize(Hsmt_tyrod,(300,300),interpolation=cv2.INTER_AREA)
thyroid2pic_reshaped = cv2.resize(thyroid2pic,(300,300),interpolation=cv2.INTER_AREA)

print('Showing Hsmt_tyrod size reshaped', Hsmt_tyrod_reshaped.shape)
print('Showing Hsmt_tyrod size reshaped', thyroid2pic_reshaped.shape)
```

```python
#print image 1
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.imshow(Hsmt_tyrod_reshaped)

#print image 2
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.imshow(thyroid2pic_reshaped)
```

```python
#image 1
#print RGB Channel
plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(Hsmt_tyrod_reshaped[:,:,0])
#insert title
plt.gca().set_title('Red Channel')

plt.subplot(132)
plt.imshow(Hsmt_tyrod_reshaped[:,:,1])
plt.gca().set_title('Green Channel')

plt.subplot(133)
plt.imshow(Hsmt_tyrod_reshaped[:,:,1])
plt.gca().set_title('Blue Channel')

#image 2
plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(thyroid2pic_reshaped[:,:,0])

plt.gca().set_title('Red Channel')

plt.subplot(132)
plt.imshow(thyroid2pic_reshaped[:,:,1])
plt.gca().set_title('Green Channel')

plt.subplot(133)
plt.imshow(thyroid2pic_reshaped[:,:,1])
plt.gca().set_title('Blue Channel')

#using the first picture as an example we may be able to clasify the image. the presense of little white dots on the bottom part may be sospecious, and also there may be a nodule on the right 
```
