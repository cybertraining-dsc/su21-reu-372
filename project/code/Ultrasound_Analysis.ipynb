```python
#3 different methods to blur images
img_blur = cv2.blur(thyroid2pic,(41,41))
img_filter = cv2.boxFilter(thyroid2pic, -5,(30,30), normalize = True)
img_Gaussianfilter = cv2.GaussianBlur(thyroid2pic,(41,41),cv2.BORDER_DEFAULT)

plt.figure(figsize=(20,20))
plt.subplot(141)
plt.imshow(thyroid2pic)
plt.gca().set_title('Original Image')
plt.axis('off')
plt.subplot(142)
plt.imshow(img_blur)
plt.gca().set_title('blur Image with blur method')
plt.axis('off')

plt.subplot(143)
plt.imshow(img_filter)
plt.gca().set_title('blur Image with boxFilter method')
plt.axis('off')

plt.subplot(144)
plt.imshow(img_Gaussianfilter)
plt.gca().set_title('blur Image with Gaussianfilter method')
plt.axis('off')
```
